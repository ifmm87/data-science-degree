```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#from apyori import apriori
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
store_data = pd.read_csv('store_data.csv', header=None)
```

```python
type(store_data)
```

```python
store_data.head()
```

```python
store_data.shape
```

```python
store_data = store_data.stack()
store_data = store_data.reset_index()
del store_data['level_1']
store_data.rename(columns={0:'items'}, inplace=True)
store_data.rename(columns={'level_0':'nro'}, inplace=True)
```

```python
store_data.head()
```

```python
store_data = store_data.groupby(['nro', 'items']).size()
```

```python
store_data.head(40)
```

```python
temp = store_data.unstack(level=-1)
```

```python
temp.head()
```

```python
productos = temp.reset_index().fillna(0).set_index('nro')
```

```python
productos.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
store_prods = productos.applymap(encode_units)
store_prods
```

```python
frequent_itemsets = apriori(store_prods, min_support=0.0053, use_colnames=True) # min_support = 0.07
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
```
