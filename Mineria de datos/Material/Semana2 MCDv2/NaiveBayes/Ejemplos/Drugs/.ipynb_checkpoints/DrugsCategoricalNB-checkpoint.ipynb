```python
import numpy as np
import pandas as pd

#import urllib.request

import sklearn
from mixed_naive_bayes import MixedNB
#from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split

from sklearn import metrics
from sklearn.metrics import accuracy_score

# Import LabelEncoder
from sklearn import preprocessing
```

```python
dataset = pd.read_csv('Drugs.csv')
```

```python
dataset.shape
```

```python
dataset.head(20)
```

```python
sex = np.array(dataset.iloc[:,1])
bp = np.array(dataset.iloc[:,2])
chol = np.array(dataset.iloc[:,3])
drg = np.array(dataset.iloc[:,6])
```

```python
drg
```

```python
label_encoder = preprocessing.LabelEncoder()
dataset['sex'] = label_encoder.fit_transform(sex)
dataset['bp1'] = label_encoder.fit_transform(bp)
dataset['chol'] = label_encoder.fit_transform(chol)
dataset['drg'] = label_encoder.fit_transform(drg)
del dataset['Sex'], dataset['BP'], dataset['Cholesterol'], dataset['Drug']
```

```python
dataset
```

```python
dataset.info()
```

```python
#X = pd.array(dataset.iloc[:,0:5])
X = dataset.iloc[0:150, 0:6].to_numpy()
y = dataset.iloc[0:150, 6].to_numpy()
```

```python
X
```

```python
y
```

```python
#X = pd.array(dataset.iloc[:,0:5])
X_test = dataset.iloc[151:, 0:6].to_numpy()
y_test = dataset.iloc[151:, 6].to_numpy()
```

```python
clf = MixedNB(categorical_features=[3,4,5])
clf.fit(X,y)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
