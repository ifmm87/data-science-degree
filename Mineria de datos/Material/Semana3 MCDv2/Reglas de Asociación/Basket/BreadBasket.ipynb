```python
df = df.groupby(['Transaction', 'Item']).size()
```

```python
df.head(20)
```

```python
temp = df.unstack(level=-1)
```

```python
basket = temp.reset_index().fillna(0).set_index('Transaction')
```

```python
basket.head()
```

```python
basket.columns
```

```python
basket.loc[2,'Scandinavian']
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True) # min_support = 0.07
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
```
